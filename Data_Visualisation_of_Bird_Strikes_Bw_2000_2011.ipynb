- A bird strike is strictly defined as a collision between a bird and an aircraft which is in
flight or on a take-off or landing roll. The term is often expanded to cover other wildlife
strikes - with bats or ground animals. Bird Strike is common and can be a significant
threat to aircraft safety. For smaller aircraft, significant damage may be caused to the
aircraft structure and all aircraft, especially jet-engine ones, are vulnerable to the loss
of thrust which can follow the ingestion of birds into engine air intakes. This has
resulted in several fatal accidents.

- Bird strikes may occur during any phase of flight, but are most likely during the
take-off, initial climb, approach and landing phases due to the greater numbers of birds
in flight at lower levels. To have a closer look the following document visually depicts
the data collected on Bird Strikes by FAA between 2000-2011.

### Visuals depicting the number of bird strikes

```python
# Count total number of bird strikes
total_bird_strikes = df.shape[0]

# Plot
plt.figure(figsize=(4, 2))
plt.text(0.5, 0.5, f"Total Bird Strikes: {total_bird_strikes:,}", ha='center', va='center', fontsize=15, color='blue')
plt.axis('off')  # Turn off axis
plt.tight_layout()
plt.show()
```

### Yearly Analysis & Bird Strikes in the US

```python
# Convert 'FlightDate' to datetime format
df['FlightDate'] = pd.to_datetime(df['FlightDate'])

# Extract year from 'FlightDate'
df['Year'] = df['FlightDate'].dt.year

# Group data by year and count bird strikes
bird_strikes_by_year = df['Year'].value_counts().sort_index()

# Plot
plt.figure(figsize=(8, 5))
plt.plot(bird_strikes_by_year.index, bird_strikes_by_year.values, marker='o', color='red', linestyle='-')
plt.title('Number of Bird Strikes in the US (Yearly Analysis)')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Drop rows with missing values in 'Aircraft: Airline/Operator' column
df.dropna(subset=['Aircraft: Airline/Operator'], inplace=True)

# Filter for US airlines (assuming 'Aircraft: Airline/Operator' contains airline names)
us_airlines = df[df['Aircraft: Airline/Operator'].str.contains('US')]

# Group data by airline and count bird strikes
top_10_airlines = us_airlines['Aircraft: Airline/Operator'].value_counts().head(10)

# Create data for funnel chart
labels = top_10_airlines.index
values = top_10_airlines.values

# Create funnel chart
fig = go.Figure(go.Funnel(
    y=labels,
    x=values,
    textinfo='value+percent initial'))

# Update layout
fig.update_layout(
    title='Top 10 US Airlines by Bird Strikes',
    xaxis_title='Number of Bird Strikes',
    yaxis_title='Airline',
    plot_bgcolor='rgba(0,0,0,0)',
    funnelmode='stack')

# Show plot
fig.show(renderer='iframe')
```

```python
top_airports = df['Airport: Name'].value_counts().head(50)
plt.figure(figsize=(14, 9))
top_airports.plot(kind='bar', color='skyblue')
plt.title('Top 50 Airports with Most Bird Strike Incidents')
plt.xlabel('Airport')
plt.ylabel('Number of Incidents')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Convert 'Cost: Total $' column to numeric
df['Cost: Total $'] = pd.to_numeric(df['Cost: Total $'], errors='coerce')

# Group by year and sum the total cost
yearly_cost = df.groupby('Year')['Cost: Total $'].sum()

yearly_cost = df.groupby('Year')['Cost: Total $'].sum()
plt.figure(figsize=(10, 6))
yearly_cost.plot(marker='o', color='green')
plt.title('Yearly Cost Incurred due to Bird Strikes (2000-2011)')
plt.xlabel('Year')
plt.ylabel('Total Cost ($)')
plt.xticks(yearly_cost.index)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
airport_counts = df['Airport: Name'].value_counts().head(10)

# Plot the top 10 airports with the most bird strike incidents
plt.figure(figsize=(10, 6))
sns.barplot(x=airport_counts.values, y=airport_counts.index, palette='viridis')
plt.title('Top 10 Airports with the Most Bird Strike Incidents')
plt.xlabel('Number of Incidents')
plt.ylabel('Airport')
plt.tight_layout()
plt.show()
```

```python
altitude_column = 'Feet above ground'
airport_column = 'Airport: Name'

# Create df_top_10 DataFrame
top_10_airports = df['Airport: Name'].value_counts().head(10).index
df_top_10 = df[df['Airport: Name'].isin(top_10_airports)]

# Visualize the altitude distribution of bird strikes for the top 10 airports using a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_top_10, x=airport_column, y=altitude_column, palette='coolwarm')
plt.title('Altitude Distribution of Bird Strikes for Top 10 Airports')
plt.xlabel('Airport')
plt.ylabel('Altitude (feet)')
plt.yscale('log')  # Use logarithmic scale for the y-axis

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Tight layout to prevent clipping of labels
plt.tight_layout()

plt.show()
```

```python
phase_of_flight_column = 'When: Phase of flight'
altitude_column = 'Feet above ground'

# Visualize the average altitude of aircraft in different phases at the time of strike using a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x=phase_of_flight_column, y=altitude_column, data=df, estimator='mean', ci=None, palette='muted')
plt.title('Average Altitude of Aircraft in Different Phases at the Time of Strike')
plt.xlabel('Phase of Flight')
plt.ylabel('Average Altitude (feet)')
plt.xticks(rotation=45)
plt.yscale('log')  # Use logarithmic scale for the y-axis
plt.tight_layout()
plt.show()
```

```python
effect_on_flight_column = 'Effect: Impact to flight'

# Count the occurrences of each effect on flight
effect_counts =df[effect_on_flight_column].value_counts()

# Plotting a pie chart
plt.figure(figsize=(10, 6))
plt.pie(effect_counts, labels=effect_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Set2'))
plt.title('Effect of Bird Strikes on Flight')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.tight_layout()
plt.show()
```
